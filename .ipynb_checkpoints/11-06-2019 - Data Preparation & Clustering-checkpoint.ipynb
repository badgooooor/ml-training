### Moving Average

```python
def process_moving_average(activity, df, time_window, filter_window):
    
    data = df[df['activity'] == activity][['x-axis', 'y-axis', 'z-axis']][:time_window]
    Filter_window = filter_window
    data_rolling_mean = data.rolling(Filter_window).mean()
    #Fill missing first rows
    fill_number = Filter_window-1
    fill_data = data[:fill_number]
    data_rolling_mean.update(fill_data)
    #Combine raw data and moving AVG
    df_01 = data
    df_01["ma_x"] = data_rolling_mean["x-axis"]
    df_01["ma_y"] = data_rolling_mean["y-axis"]
    df_01["ma_z"] = data_rolling_mean["z-axis"]
    #Plotting
    fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(16, 12))
    ax[0].set_title(activity)
    ax[0].plot(df_01[["x-axis"]], label='x-axis',color="b")
    ax[0].plot(df_01[["ma_x"]], label='ma_x',color="red")
    ax[0].legend(loc='lower left', bbox_to_anchor=(1.0, 0.5))
    
    ax[1].plot(df_01[["y-axis"]], label='y-axis',color="orange")
    ax[1].plot(df_01[["ma_y"]], label='ma_y',color="red")
    ax[1].legend(loc='lower left', bbox_to_anchor=(1.0, 0.5))
    
    ax[2].plot(df_01[["z-axis"]], label='z-axis',color="g")
    ax[2].plot(df_01[["ma_z"]], label='ma_z',color="red")
    ax[2].legend(loc='lower left', bbox_to_anchor=(1.0, 0.5))
    return df_01
```

```python
process_moving_average('Walking', df, 100, 20)
```

```python
# Import and get raw data(get only value from axis)
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
df_raw = df[['x-axis','y-axis','z-axis']].iloc[0:10000,:]
df_raw
```

### Raw data -> Min-max scaler

```python
# normalized raw data with Min-max scaler
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df_raw)
print('Minimum value in each axis : ',scaler.data_min_)
print('Maximum value in each axis : ',scaler.data_max_)
df_scaled = pd.DataFrame(data=df_scaled, columns=['x-axis','y-axis','z-axis'])
df_scaled.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df_scaled['x-axis'],df_scaled['y-axis'],df_scaled['z-axis'])
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.show()
```

```python
def clustering(data):
# K-mean clustering
    kmeans = KMeans(n_clusters=3, random_state=0).fit(data)
    print('Center : ',kmeans.cluster_centers_)

# Plot k-mean result
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    labels = kmeans.labels_
    ax.scatter(data['x-axis'],data['y-axis'],data['z-axis'],c=labels.astype(np.float),edgecolor='k')
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_xlabel('x-axis')
    ax.set_ylabel('y-axis')
    ax.set_zlabel('z-axis')
    plt.show()
clustering(df_scaled)
```

### Moving average

```python
# Moving average
def process_moving_average(df, time_window, filter_window):
    
    data = df[['x-axis', 'y-axis', 'z-axis']][:time_window]
    Filter_window = filter_window
    data_rolling_mean = data.rolling(Filter_window).mean()
    #Fill missing first rows
    fill_number = Filter_window-1
    fill_data = data[:fill_number]
    data_rolling_mean.update(fill_data)
    #Combine raw data and moving AVG
    df_01 = data
    df_01["ma_x"] = data_rolling_mean["x-axis"]
    df_01["ma_y"] = data_rolling_mean["y-axis"]
    df_01["ma_z"] = data_rolling_mean["z-axis"]
    #Plotting
    fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(16, 12))
    ax[0].plot(df_01[["x-axis"]], label='x-axis',color="b")
    ax[0].plot(df_01[["ma_x"]], label='ma_x',color="red")
    ax[0].legend(loc='lower left', bbox_to_anchor=(1.0, 0.5))
    
    ax[1].plot(df_01[["y-axis"]], label='y-axis',color="orange")
    ax[1].plot(df_01[["ma_y"]], label='ma_y',color="red")
    ax[1].legend(loc='lower left', bbox_to_anchor=(1.0, 0.5))
    
    ax[2].plot(df_01[["z-axis"]], label='z-axis',color="g")
    ax[2].plot(df_01[["ma_z"]], label='ma_z',color="red")
    ax[2].legend(loc='lower left', bbox_to_anchor=(1.0, 0.5))
    return df_01
print('Processed data with moving average')
df_ma_raw = process_moving_average(df_raw, 500, 5)
df_ma_raw.head()
```

```python
# Replace moving average axis value
df_ma = df_ma_raw[['ma_x','ma_y','ma_z']]
df_ma.columns = ['x-axis','y-axis','z-axis']
df_ma
```
